> **Read_data Function Do**

1.   Input /paramart the path of file
2.   return date as dataframe






```
def read_data(url):
  try:
    df= pd.read_csv(url,error_bad_lines=False)
  except FileNotFoundError:
    return('FileNotFoundError') 
  return(df)
```



> **reomve_highly_corr Function Do**

1.   Input /paramart the path of file
2.   remove corr more than .85






```
def reomve_highly_corr(df):
  # Calculate the correlation matrix and take the absolute value
  p_corr = df.corr(method='pearson').abs()

  # Create a True/False mask and apply it
  mask = np.triu(np.ones_like(p_corr, dtype=bool))
  tri_df = p_corr.mask(mask)

  # List column names of highly correlated features (P_corr > 0.95)
  to_drop = [c for c in tri_df.columns if any(tri_df[c] >  0.85)]

  # Drop the features in the to_drop list
  reduced_df = df.drop(to_drop, axis=1)
  return(reduced_df)
```

**Runcode Function Do**

> Run all previous function. 

> Take one paramert Path of file. 





```
def runcode(url):
  df =read_data(url)
  df_change=reomve_highly_corr(df)
  return(df_change)
```
