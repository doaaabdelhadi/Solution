> **Read_data Function Do**

1.   Input /paramart the path of file
2.   return date as dataframe






```
def read_data(url):
  try:
    df= pd.read_csv(url,error_bad_lines=False)
    
  except FileNotFoundError:
    return('FileNotFoundError') 
  return(df)   
  
```

 
**Check_dates_col Function do**

1. paramrt/ Input(dataframe which return from read_data function). 
2. Check if a input is dataframe or not. if not return there not dataframe.
3. if input is dataframe check if object try to convert to data. 
if convert that mean the colume type is data. 
4. cleck the data colume only and put dataframe. 


```
def check_dates_col(df):
  if isinstance(df, pd.DataFrame):
    for col in df.columns:
      if df[col].dtype == 'object':
          try:
              df[col] = pd.to_datetime(df[col])
          except ValueError:
              pass
    dates_df =df.select_dtypes(include=[np.datetime64]) 
    count=0
    col_name=[]
    for item in dates_df.columns:
      count=count+1
      col_name.append('date{}'.format(count))
    dates_df.columns=col_name  
    return(dates_df)

  return('This is not vaild please check your file')
```

**Function do**

1. Paramet / input dataframe from (check_dates_col)
2.   if len of list of columne more than one run difference function.
     if not return data frame of dates only.


> **Option**
apply absolute function (removes the minus sign (-)) 





```
def difference(df):
    col=df.columns
    diff_df = pd.DataFrame()
    if len(col) >1:
      for i in range(len(col)):
        diff_df['{}'.format(col[i-1]+'-'+col[i])]=(df[col[i-1]].sub(df[col[i]], axis=0))/ np.timedelta64(1, 'D')
      #option 
      # if you want to test only uncomment the next line and comment the previous line.
      # diff_df['{}'.format(col[i-1]+'-'+col[i])]=(df[col[i-1]].sub(df[col[i]], axis=0)).abs()/ np.timedelta64(1, 'D')
      return(diff_df)  
    else:
      return(df)

    
```

**Runcode Function Do**

> Run all previous function. 

> Take one paramert Path of file. 





```
def runcode(url):
  df=read_data(url)
  dates=check_dates_col(df)
  col=dates.columns
  diff_dates=difference(dates)
  return(diff_dates)  
    

```
